Описание колонок:

- `instance_id` - Уникальный идентификатор трека;
- `track_name` - Название трека;
- `acousticness` - Мера уверенности от 0,0 до 1,0 в том, что трек является акустическим: 1,0 означает высокую степень уверенности в том, что трек является акустическим;
- `danceability` - Танцевальность, описывает, насколько трек подходит для танцев, основываясь на сочетании музыкальных элементов, включая темп, стабильность ритма, силу ударов и общую регулярность. Значение 0,0 означает наименьшую танцевальность, а 1,0 - наибольшую танцевальность;
- `duration_ms` - Продолжительность трека в миллисекундах;
- `energy` - Энергия, это показатель от 0,0 до 1,0, представляющий собой меру интенсивности и активности. Как правило, энергичные композиции ощущаются как быстрые, громкие и шумные. Например, дэт-метал обладает высокой энергией, в то время как прелюдия Баха имеет низкую оценку этого параметра;
- `instrumentalness` - Определяет, содержит ли трек вокал. Звуки "Ooh" и "aah" в данном контексте рассматриваются как инструментальные. Рэп или разговорные треки явно являются "вокальными". Чем ближе значение инструментальности к 1,0, тем больше вероятность того, что трек не содержит вокала;
- `key` - базовый ключ (нота) произведения;
- `liveness` - Определяет присутствие аудитории в записи. Более высокие значения liveness означают увеличение вероятности того, что трек был исполнен вживую. Значение выше 0,8 обеспечивает высокую вероятность того, что трек исполняется вживую;
- `loudness` - Общая громкость трека в децибелах (дБ);
- `mode` - Указывает на модальность (мажорную или минорную) трека;
- `speechiness` - Речевой характер определяет наличие в треке разговорной речи. Чем более исключительно речевой характер носит запись (например, ток-шоу, аудиокнига, поэзия), тем ближе значение атрибута к 1,0. Значения выше 0,66 характеризуют треки, которые, вероятно, полностью состоят из разговорной речи. Значения от 0,33 до 0,66 характеризуют треки, которые могут содержать как музыку, так и речь, как в виде фрагментов, так и в виде слоев, включая такие случаи, как рэп-музыка. Значения ниже 0,33, скорее всего, представляют музыку и другие неречевые треки;
- `tempo` - Темп трека в ударах в минуту (BPM). В музыкальной терминологии темп представляет собой скорость или темп данного произведения и напрямую зависит от средней продолжительности тактов;
- `obtained_date` - дата загрузки в сервис;
- `valence` - Показатель от 0,0 до 1,0, характеризующий музыкальный позитив, передаваемый треком. Композиции с высокой валентностью звучат более позитивно (например, радостно, весело, эйфорично), а композиции с низкой валентностью - более негативно (например, грустно, депрессивно, сердито);
- `music_genre` - Музыкальный жанр трека.

## Предобработка данных

### Разбиение данных на выборки

Столбцы с уникальным идентификатором, названием и даты добавления трека не влияют на жанр трека, их можно не включать в список признаков.

```python
features_train = data_train.drop(columns = ['music_genre', 'instance_id', 'track_name', 'obtained_date'])
target_train = data_train['music_genre'].to_frame(name='music_genre')

print(
    'Размеры features_train:', 
    features_train.shape,
    'Размеры target_train:', 
    target_train.shape,
    sep='\n'
)
```

### Отбор колонок

```python
cat_columns = features_train.select_dtypes(exclude='number').columns.to_list()
cat_columns
```

```python
num_columns = features_train.select_dtypes(include='number').columns.to_list()
num_columns
```

### Обзор качественных переменных

```python
for cat in cat_columns:
    print(
        f'Колонка {cat} features_train:', 
        features_train[cat].unique().tolist(),
        f'Колонка {cat} features_test:',
        features_test[cat].unique().tolist(),
        sep='\n'
    )
    print()
```

В обоих выборках содержатся одинаковые значения в одинаковой записи.

### Обзор кол-ных переменных

```python
features_train[num_columns].describe()
```

```python
features_train.duration_ms[features_train.duration_ms == -1].count()
```

```python
features_test.duration_ms[features_test.duration_ms == -1].count()
```

В столбце с длительностью трека как в обучающей так и в тестовой выборке есть значение -1, таких 2009 и 509 соответственно, скорее всего они означают пропуски. Заменим -1 на значение NaN для дальнейшего заполнения пропусков.

```python
features_train.loc[features_train.duration_ms == -1, 'duration_ms'] = np.nan
```

```python
features_test.loc[features_test.duration_ms == -1, 'duration_ms'] = np.nan
```

### Обзор пропусков

```python
features_train.isna().sum()
```

```python
features_test.isna().sum()
```

**Вывод:** в обучающих и тестовых данных есть пропуски в одинаковых столбцах, в 2 категориальных и в 1 кол-ном. Те же самые пропуски и в итоговой тестовой выборке. Пропуски в категориальных столбцах можно заменить на значение `unknown`, а в числовых - с помощью класса `IterativeImputer`, который моделирует каждую функцию с пропущенными значениями в зависимости от других функций и использует эту оценку для восстановления.

### Заполнение пропусков

```python
col_transformer_nan = make_column_transformer(
    (SimpleImputer(strategy = 'constant', fill_value = 'unknown'), cat_columns),
    (IterativeImputer(max_iter=20, random_state = RANDOM_STATE), num_columns),
    remainder='passthrough',
    verbose_feature_names_out=False
)
```

```python
features_train = pd.DataFrame(
    col_transformer_nan.fit_transform(features_train),
    columns=col_transformer_nan.get_feature_names_out()
)

features_train[num_columns] = features_train[num_columns].astype(float)
get_info_data(features_train)
```

```python
features_test = pd.DataFrame(
    col_transformer_nan.transform(features_test),
    columns=col_transformer_nan.get_feature_names_out()
)

features_test[num_columns] = features_test[num_columns].astype(float)
get_info_data(features_test)
```

## Проверка на мультиколлинеарность и создание новых признаков

```python
def get_corr_matrix(data, columns):
    corr_matrix = data[columns].phik_matrix()

    plot_correlation_matrix(
        corr_matrix.values, 
        x_labels=corr_matrix.columns, 
        y_labels=corr_matrix.index, 
        vmin=0, 
        vmax=1, 
        color_map='Blues', 
        title='Матрица корреляции', 
        fontsize_factor=0.8, 
        figsize=(10, 8)
    )
```

```python
def get_high_corr_matrix(data, columns):
    corr_matrix = data[columns].phik_matrix()
    
    no_md_matrix = pd.DataFrame(corr_matrix.mask(np.eye(corr_matrix.shape[0], dtype = bool)), columns=corr_matrix.columns)
    corr_features = [col for col in no_md_matrix.columns if no_md_matrix[col].max() >= 0.7]
    high_corr_matrix = corr_matrix.loc[corr_features][corr_features]

    plot_correlation_matrix(
        high_corr_matrix.values, 
        x_labels=high_corr_matrix.columns, 
        y_labels=high_corr_matrix.index, 
        vmin=0, 
        vmax=1, 
        color_map='Blues', 
        title='Матрица высококоррелирующих признаков', 
        fontsize_factor=1, 
        figsize=(5, 3)
    )
```

```python
data_train_new = pd.concat([features_train, target_train], axis=1)
matrix_column = data_train_new.columns.tolist()
```

```python
get_corr_matrix(data_train_new, matrix_column)
```

На ключевой признак оказывают особое влияние `loudness`, `energy`, `acousticness` и `danceability`.

```python
get_high_corr_matrix(data_train_new, matrix_column)
```

Между собой находятся в большой зависимости 3 признака: `loudness`, `energy` и `acousticness`, два из них нужно удалить для корректной работы модели.

Так как на жанр трека больше всего влияет `loudness`, то оставим его.

## Отбор финального набора обучающих признаков

Итак, в финальный набор признаков войдут:

- `danceability` - Танцевальность, описывает, насколько трек подходит для танцев, основываясь на сочетании музыкальных элементов, включая темп, стабильность ритма, силу ударов и общую регулярность. Значение 0,0 означает наименьшую танцевальность, а 1,0 - наибольшую танцевальность;
- `duration_ms` - Продолжительность трека в миллисекундах;
- `instrumentalness` - Определяет, содержит ли трек вокал. Звуки "Ooh" и "aah" в данном контексте рассматриваются как инструментальные. Рэп или разговорные треки явно являются "вокальными". Чем ближе значение инструментальности к 1,0, тем больше вероятность того, что трек не содержит вокала;
- `key` - базовый ключ (нота) произведения;
- `liveness` - Определяет присутствие аудитории в записи. Более высокие значения liveness означают увеличение вероятности того, что трек был исполнен вживую. Значение выше 0,8 обеспечивает высокую вероятность того, что трек исполняется вживую;
- `loudness` - Общая громкость трека в децибелах (дБ);
- `mode` - Указывает на модальность (мажорную или минорную) трека;
- `speechiness` - Речевой характер определяет наличие в треке разговорной речи. Чем более исключительно речевой характер носит запись (например, ток-шоу, аудиокнига, поэзия), тем ближе значение атрибута к 1,0. Значения выше 0,66 характеризуют треки, которые, вероятно, полностью состоят из разговорной речи. Значения от 0,33 до 0,66 характеризуют треки, которые могут содержать как музыку, так и речь, как в виде фрагментов, так и в виде слоев, включая такие случаи, как рэп-музыка. Значения ниже 0,33, скорее всего, представляют музыку и другие неречевые треки;
- `tempo` - Темп трека в ударах в минуту (BPM). В музыкальной терминологии темп представляет собой скорость или темп данного произведения и напрямую зависит от средней продолжительности тактов;
- `valence` - Показатель от 0,0 до 1,0, характеризующий музыкальный позитив, передаваемый треком. Композиции с высокой валентностью звучат более позитивно (например, радостно, весело, эйфорично), а композиции с низкой валентностью - более негативно (например, грустно, депрессивно, сердито).

```python
cat_features = ['mode', 'key']

num_features = [
    'danceability', 
    'duration_ms', 
    'instrumentalness', 
    'liveness', 
    'loudness', 
    'speechiness',
    'valence',
    'tempo'
]

all_columns = cat_features + num_features
```

## Выбор и обучение моделей

В качестве моделей буду использовать:

- `RidgeClassifier`;
- `RandomForestClassifier`;
- `CatBoostClassifier`.

Так как предсказание жанра трека - это задача мультиклассификации, в роли метрики для сравния качества моделей буду использовать `Micro F1 score`. Эта метрика, в отличие от обычной формулы `F1 score`, рассчитывается с использованием общего количества истинно положительных результатов (TP), ложноположительных результатов (FP) и ложноотрицательных результатов (FN), а не индивидуально для каждого класса.

Перед обучением числовые данные необходимо стандартизировать, а категориальные - масштабировать.

```python
col_transformer_processing = make_column_transformer(
    (OneHotEncoder(drop='first', handle_unknown='ignore'), cat_features),
    (StandardScaler(), num_features),
    remainder='passthrough',
    verbose_feature_names_out=False
)
```

**Model RidgeClassifier**

```python
paramsRidge = {'ridgeclassifier__alpha': range(1, 10)}

modelRidge = RidgeClassifier(
    random_state=RANDOM_STATE, 
    alpha=1,
    class_weight='balanced'
)

pipelineRidge = make_pipeline(col_transformer_processing, modelRidge)
pipelineRidge
```

```python
gridsearch_Ridge = GridSearchCV(pipelineRidge, paramsRidge, cv=5, scoring='f1_micro')
```

```python
gridsearch_Ridge.fit(features_train[all_columns], target_train['music_genre'])
```

```python
print('Best Parameters:', gridsearch_Ridge.best_params_)
print('Best Score:', gridsearch_Ridge.best_score_)
```

**Model RandomForestClassifier**

```python
paramsRFC = {
    'randomforestclassifier__n_estimators': range(10, 201, 10),
    'randomforestclassifier__max_depth': range(1, 21),
    'randomforestclassifier__min_samples_leaf': range(2, 11),
    'randomforestclassifier__min_samples_split': range(2, 11)
}

modelRFC = RandomForestClassifier(
    random_state = RANDOM_STATE, 
    n_estimators=50, 
    max_depth=5, 
    min_samples_leaf=4, 
    min_samples_split=4, 
    class_weight='balanced'
)

pipelineRFC = make_pipeline(col_transformer_processing, modelRFC)
pipelineRFC
```

```python
randomsearch_RFC = RandomizedSearchCV(
    pipelineRFC, 
    param_distributions=paramsRFC, 
    n_iter=50,
    cv=5, 
    scoring='f1_micro',
    random_state=RANDOM_STATE
)
```

```python
randomsearch_RFC.fit(features_train[all_columns], target_train['music_genre'])
```

```python
print('Best Parameters:', randomsearch_RFC.best_params_)
print('Best Score:', randomsearch_RFC.best_score_)
```

**Model CatBoostClassifier**

```python
paramsCDC = {
    'catboostclassifier__iterations': range(100, 1000, 100),  # Количество итераций
    'catboostclassifier__depth': range(1, 11),  # Глубина деревьев
    'catboostclassifier__learning_rate': uniform(0.01, 0.3),  # Скорость обучения
    'catboostclassifier__l2_leaf_reg': uniform(1, 10),  # L2-регуляризация
    'catboostclassifier__border_count': range(1, 255),  # Количество точек разбиения
}

modelCDC = CatBoostClassifier(
    loss_function='MultiClass', 
    auto_class_weights='Balanced', 
    random_seed=RANDOM_STATE,
    verbose=False
)

pipelineCDC = make_pipeline(col_transformer_processing, modelCDC)
pipelineCDC
```

```python
randomsearch_CDC = RandomizedSearchCV(
    pipelineCDC, 
    param_distributions=paramsCDC, 
    n_iter=40,
    cv=5, 
    scoring='f1_micro',
    random_state=RANDOM_STATE
)
```
